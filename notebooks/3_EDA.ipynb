"Exploratory Data Analysis (EDA)"
    Objective--
        Understand customer behavior patterns, identify churn drivers, and validate data quality before machine learning modeling.

EDA focuses on relationships between customer attributes and churn outcomes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/Cleaned_Telco_Churn.csv")
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

-Churn Distribution Analysis-
    The dataset shows moderate class imbalance, with non-churn customers significantly outnumbering churn customers.
    This indicates that evaluation metrics such as Precision, Recall, and F1-score will be more appropriate than accuracy alone during modeling.

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
```

-Monthly Charges vs Churn-
    Monthly charges show a mild positive relationship with churn risk. Customers with higher monthly charges show slightly increased churn likelihood.
    This may indicate price sensitivity or dissatisfaction with higher-cost plans.

```python
sns.boxplot(x='Churn', y='TotalCharges', data=df)
```

-Total Charges vs Churn-
    Several high TotalCharges values appear as statistical outliers in churned customers. These values were retained because they represent real high-value customer behavior rather than data errors.

```python
sns.boxplot(x='Churn', y='tenure', data=df)
```

-Tenure vs Churn-
    Analysis of tenure shows that customers with lower tenure are significantly more likely to churn. This indicates that customer retention risk is highest in early service periods
    Tenure is expected to be one of the strongest predictive features.

```python
sns.countplot(x='Contract', hue='Churn', data=df)
```

-Contract Type vs Churn-
    Customers with month-to-month contracts show significantly higher churn rates compared to long-term contract customers.
    Long-term contracts appear to improve customer retention stability.

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True)
```

Key EDA Insights
    -Customers with shorter tenure are more likely to churn.
    -Month-to-month contract customers have the highest churn probability.
    -Higher monthly charges show mild positive correlation with churn risk.
    -TotalCharges strongly correlates with tenure and customer lifetime value.
    -Churn dataset shows moderate class imbalance.
